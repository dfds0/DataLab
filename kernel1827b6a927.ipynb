# Titanic: Machine Learning from Disaster

```python
# Imports
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load of files
gender_submission = pd.read_csv("../input/titanic/gender_submission.csv")
test = pd.read_csv("../input/titanic/test.csv")
train = pd.read_csv("../input/titanic/train.csv")
```

```python
train.head()
# gender_submission.head()
# test.head()
```

```python
# Define a numeric value for the sex
def convert_sex(value):
    if value == 'female':
        return 1
    else:
        return 0

# Append a new column with the numeric value for the sex
train['Sex_binary'] = train['Sex'].map(convert_sex)
test['Sex_binary'] = test['Sex'].map(convert_sex)

train.head()
```

```python
# Build model
model = RandomForestClassifier(n_estimators=100, n_jobs=1, random_state=0)
```

```python
# Selection of features and target
X_train = train[['Sex_binary', 'Age']]
y_train = train['Survived']

# Append default value
X_train = X_train.fillna(-1)

# Fit the model
model.fit(X_train, y_train)
```

```python
# Selection of features and target
X_test = test[['Sex_binary', 'Age']]

# Append default value
X_test = X_test.fillna(-1)

# Predict the Y_test
Y_test = model.predict(X_test)
```

```python
# Convert in Series
Y_test = pd.Series(Y_test, index=test['PassengerId'], name="Survived")

# Write a csv file
Y_test.to_csv("first_model.csv", header=True)
```
