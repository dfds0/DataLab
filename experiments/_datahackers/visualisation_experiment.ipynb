```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pandas.plotting import autocorrelation_plot
from pandas.plotting import scatter_matrix
from pandas.plotting import parallel_coordinates
from pandas.plotting import lag_plot
```

```python
ds_car = sns.load_dataset('mpg')
ds_meal = sns.load_dataset('tips')
ds_attetion = sns.load_dataset('attention')
```

---

```python
plt.scatter(ds_car.acceleration, ds_car.horsepower, marker='^')
plt.show();
```

```python
ds_car.plot.hexbin(x='acceleration', y='horsepower', gridsize=10, cmap='BuPu_r')
plt.show();
```

```python
sns.heatmap(ds_car.corr(), annot=True, cmap='PiYG_r')
plt.show();
```

```python
autocorrelation_plot(ds_meal.total_bill)
plt.show();
```

```python
parallel_coordinates(ds_attetion, 'attention', color=('red', 'blue'))
plt.show();
```
