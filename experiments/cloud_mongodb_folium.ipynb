```python
import pymongo
import pprint
import urllib.parse
import uuid
import pandas as pd
import folium
```

```python
client = pymongo.MongoClient("mongodb+srv://dbuser:dbpassword@cluster0-tdmr9.mongodb.net/test?retryWrites=true&w=majority")
```

```python
# Load the database
database = client['sample_geospatial']

# Load a collection
shipwrecks = database['shipwrecks']

# Load a instance of Shipwrecks (Don't use all)
instances = shipwrecks.find().sort("_id",pymongo.DESCENDING)[:1000]
```

```python
dataframe = pd.DataFrame(instances)
dataframe.head()
```

```python
map_instance = folium.Map(zoom_start=5)

ships = 0
for index, row in dataframe.iterrows():
    latdec = row['latdec']
    londec = row['londec']
    folium.map.Marker([latdec, londec]).add_to(map_instance)    
    #print('lat:', latdec, ' lon:', londec)
    ships += 1

map_instance.add_child(folium.LatLngPopup())

print("Number of points:", ships)
```
