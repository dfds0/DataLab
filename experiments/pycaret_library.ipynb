---

### PyCaret: Settings

```python
import pandas as pd
from pycaret import classification 
```

---

### PyCaret: DataFrame

```python
df_classification = pd.read_csv('https://raw.githubusercontent.com/lakshay-arora/PyCaret-Example/master/datasets/loan_train_data.csv')
df_classification.head()
```

---

### PyCaret: Classification Setup

```python
classification_setup = classification.setup(data=df_classification, target='Personal Loan')

classification_setup;
```

---

### PyCaret: Model - Decision Tree

```python
# Create a model for decision_tree = 'dt'
model_decision_tree = classification.create_model('dt')

model_decision_tree
```

```python
# Evaluate
classification.evaluate_model(model_decision_tree)
```

```python
# Saving the model
classification.save_model(model_decision_tree, 'model_decision_tree')
```

---

### PyCaret: Model - XGBoost

```python
# Create a model for XGBoost = 'xgboost'
model_xgboost = classification.create_model('xgboost')

model_xgboost
```

```python
# Evaluate
classification.evaluate_model(model_xgboost)
```

```python
# Saving the model
classification.save_model(model_xgboost, 'model_xgboost')
```

---
